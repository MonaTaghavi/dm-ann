Load Iris data and visualize it

```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
import pickle
```

```python
data = load_iris()
df = pd.DataFrame(data=data.data, columns= data.feature_names)
df.head()
```

```python
df["Species"]= data.target
df["Species_Name"] = df["Species"].map(lambda species: data.target_names[species])
df.head()
```

```python
df.head(70)
```

Check missing values

```python
df.isna().sum()
```

```python
df = df.astype({"Species": "category"})
df.dtypes
```

```python
df.describe()
```

```python
#Histogram
fig = px.histogram(df, x="Species")
fig.show()
```

```python
#Boxplot
fig = px.box(df, x=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
                   ,color="Species_Name")
fig.show()
```

```python
fig = px.violin(df, y=["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
                   ,color="Species_Name", points="all", box=False)
fig.show()
```

```python
#Scatter plot
fig = px.scatter(df, x="sepal length (cm)", y="sepal width (cm)", color="Species_Name")
fig.show()
```

```python
fig = px.scatter(df, x="petal length (cm)", y="petal width (cm)", color="Species_Name")
fig.show()
```

```python
fig = px.scatter_3d(df, x="petal length (cm)", y="petal width (cm)", z= "sepal length (cm)", 
                    color="Species_Name", height=500)
fig.show()
```

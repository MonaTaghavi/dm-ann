Load Iris data and visualize it

```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import pickle
```

```python
data = load_iris()
df = pd.DataFrame(data=data.data, columns= data.feature_names)
df.head()
```

```python
df["Species"]= data.target
df["Species_Name"] = df["Species"].map(lambda species: data.target_names[species])
df.head()
```

Standard Scaler

```python
columns_to_scale = ["sepal length (cm)", "sepal width (cm)", "petal width (cm)", "petal length (cm)"]

scaler = StandardScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df.describe()
```

Train / Test Split

```python
x = df.drop(columns=["Species", "Species_Name"])
y = df["Species"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f"Train Shape: {x_train.shape}")
print(f"Test Shape: {x_test.shape}")
```

Neural Network Classifier

```python
model = MLPClassifier(random_state=42, max_iter=500, hidden_layer_sizes=(50, 10 ), 
                      n_iter_no_change=50, early_stopping=True, verbose=True).fit(x_train, y_train)
```

```python
loss_curve = model.loss_curve_
plt.plot(loss_curve)
plt.show()
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
validation_scores = model.validation_scores_
plt.plot(validation_scores)
plt.show()
```

```python
predictions = model.predict(x_test)
cm = confusion_matrix(y_test, predictions, labels=model.classes_)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
```

```python
import pandas as pandas
import numpy as np
from sklearn.neural_network import MLPClassifier
from matplotlib import pyplot as plt
```

```python
data = {"a": [0, 0, 1, 1], "b": [0, 1, 0, 1], "y": [0, 1, 1, 0]}
df = pandas.DataFrame(data)
df
```

```python
train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]).reshape(4, 2)
target = np.array([0, 1, 1, 0])
print(train)
print(target)
```

```python
model = MLPClassifier(random_state=42, max_iter=3000, hidden_layer_sizes=(6,))
model.fit(train, target)
```

```python
model.score(train, target)
```

```python
loss_values = model.loss_curve_
plt.plot(loss_values)
plt.show()
```

```python
test = np.array([[0, 1]])
print(model.predict(test))
```

```python
model_weights = model.coefs_
print(model_weights)
```

```python
model_biases = model.intercepts_
print(model_biases)
```

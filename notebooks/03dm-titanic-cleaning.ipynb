```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_titanic = pd.read_csv("../data/titanic.csv")
df_titanic.head()
```

```python
df_titanic.info()
```

```python
df_titanic.isna().sum()
```

```python
df_titanic.shape
```

```python
df_titanic.dropna(subset=["Embarked"], inplace=True)
```

```python
df_titanic.shape
```

```python
df_titanic.drop(columns=["Cabin"], inplace=True)
```

```python
df_titanic.shape
```

```python
df_titanic.isna().sum()
```

```python
df_titanic.describe()
```

```python
df_titanic.hist("Age", bins=30)
plt.show()
```

```python
plt.hist(df_titanic["Age"].dropna(), bins=30, edgecolor="black", color="skyblue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Ages")
plt.show()
```

```python
df_temp = df_titanic.copy()
```

```python
df_temp.fillna(value={"Age":0}, inplace=True)
```

```python
df_temp.isna().sum()
```

```python
df_temp.describe()
```

```python
plt.hist(df_temp["Age"], bins=30, edgecolor="black", color="skyblue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Ages")
plt.show()
```

```python
df_temp = df_titanic.copy()
df_temp.fillna(value={"Age": df_temp["Age"].mean()}, inplace=True)
```

```python
plt.hist(df_temp["Age"], bins=30, edgecolor="black", color="skyblue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Ages")
plt.show()
```

```python
df_temp.describe()
```

```python
df_temp = df_titanic.copy()
df_temp.head(10)
df_temp["Age"]= df_temp["Age"].bfill()
df_temp.head(10)
```

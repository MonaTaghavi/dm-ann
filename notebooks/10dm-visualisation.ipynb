Matplotlib Vs Plotly

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_loan = pd.read_csv("../data/loan_nomissing.csv")
df_loan.head()
```

```python
df_loan.dtypes
```

Pandas

```python
df_loan["ApplicantIncome"].hist(bins=50)
```

Matplotlib

```python
plt.hist(df_loan["ApplicantIncome"], bins=50)
plt.show()
```

```python
plt.scatter(df_loan["ApplicantIncome"], df_loan["Gender"])
plt.show()
```

Plotly

```python
import plotly.express as px
import numpy as np
```

```python
fig = px.histogram(df_loan, x="ApplicantIncome", nbins=50)
fig.show()
```

```python
fig = px.scatter(df_loan, x="ApplicantIncome", y="CoapplicantIncome", color="Gender",
                 labels={"ApplicantIncome": "Applicant Income", "CoapplicantIncome": "Coapplicant Income"},
                 title="Applicant Income vs Coapplicant Income")
fig.show()
```

Titanic- 3D plot

```python
df_titanic = pd.read_csv("../data/titanic_cleaned.csv")
df_titanic.head()
```

```python
fig = px.scatter_3d(df_titanic, x="Age", y="Fare", z="Sex", color="Survived", size="Pclass",
                    height= 1000)
fig.show()
```

Box Plot in Plotly

```python
fig = px.box(df_titanic, x="Age")
fig.show()
```

```python
fig = px.box(df_titanic, x="Age", color="Sex")
fig.show()
```

```python
fig = px.box(df_titanic, x=["Age", "Fare"], color="Sex")
fig.show()
```

Violin Plot

```python
fig = px.violin(df_titanic, x="Age", color="Sex", points= False)
fig.show()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/titanic_prepared.csv")
df.head()
```

pd.iterrows()

```python
df["Gender"] = np.nan
df.head()
```

```python
%%time

for index, row in df.iterrows():
    if row["Sex"] == "male":
        df.loc[index, "Gender"] = 1
    else:
        df.loc[index, "Gender"] = 0
```

pd.itertuples()

```python
df["Gender"] = None
df.head()
```

```python
%%time

for row in df.itertuples():
    if row.Sex == "male":
        df.loc[row.Index, "Gender"] = 1
    else:
        df.loc[row.Index, "Gender"] = 0
```

np.where()

```python
df.head()
```

```python
df["Gender"] = None
df.head()
```

```python
%%time
df["Gender"] = np.where(df["Sex"] == "male", 1, 0)
```

pd.apply()

```python
df["Gender"] = None
```

```python
%%time
def gender_encoder(x):
    if x == "male":
        return 1
    else:
        return 0

df["Gender"] = df["Sex"].apply(gender_encoder)
```

```python
df["Gender"] = None
```

```python
df["Gender"] = df["Sex"].apply(lambda x: 1 if x == "male" else 0)
```

```python
%%time

tqdm.pandas(desc="Processing rows")

def gender_encoder(x):
    if x == "male":
        return 1
    else:
        return 0

df["Gender"] = df["Sex"].progress_apply(gender_encoder)
```

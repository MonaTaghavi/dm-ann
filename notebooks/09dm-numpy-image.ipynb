```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('../data/Lenna.png')
img
```

```python
img.size
```

```python
img = img.resize((250,250))
```

```python
img
```

Creating the array of the image

```python
img_array = np.array(img)
img_array
```

```python
plt.imshow(img_array)
```

RGB Channels

```python
cmaps = ["Reds", "Greens", "Blues"]
for i, cmap in enumerate(cmaps):
    plt.imshow(img_array[:,:,i], cmap=cmap, vmin=0, vmax=255)
    plt.title(cmap)
    plt.show()
```

```python
img_array_2 = img_array.copy()
#Red channel
img_array_2[:,:,0] = img_array[:,:,0] * 0.4
#Green channel
img_array_2[:,:,1] = img_array[:,:,1] * 0.6 
#Blue channel
img_array_2[:,:,2] = img_array[:,:,2] * 0.6
plt.imshow(img_array_2)
```

Crop

```python
plt.imshow(img_array[100:200, 100:200, :])
```

Rotate

```python
plt.imshow(np.rot90(img_array))
```

```python
from scipy.ndimage import rotate
plt.imshow(rotate(img_array, 45))
plt.show()
```

GrayScale

1- Average

```python
img_gray = np.average(img_array, axis=2)
img_gray = img_array.copy()
img_gray[:,:,0] = np.average(img_array, axis=2) 
img_gray[:,:,1] = img_gray[:,:,0]
img_gray[:,:,2] = img_gray[:,:,0]
plt.imshow(img_gray)
```

```python
img_gray.shape
```

2- MinMax

```python
img_minmax = img_array.copy()
img_minmax[:,:,0] = np.max(img_array, axis=-1)/2 + np.min(img_array, axis=-1)/2
img_minmax[:,:,1] = img_minmax[:,:,0]
img_minmax[:,:,2] = img_minmax[:,:,0]
plt.imshow(img_minmax)
```

3- Luma

```python
img_luma = img_array.copy()
img_luma[:,:,0] = np.dot(img_array, [0.299, 0.587, 0.114])
img_luma[:,:,1] = img_luma[:,:,0]
img_luma[:,:,2] = img_luma[:,:,0]
plt.imshow(img_luma)
```

4- Min

```python
img_min = img_array.copy()
img_min[:,:,0] = np.min(img_array, axis=-1)
img_min[:,:,1] = img_min[:,:,0]
img_min[:,:,2] = img_min[:,:,0]
plt.imshow(img_min)
```

5- Max

```python
img_max = img_array.copy()
img_max[:,:,0] = np.max(img_array, axis=-1) 
img_max[:,:,1] = img_max[:,:,0]
img_max[:,:,2] = img_max[:,:,0]
plt.imshow(img_max)
```

Comparison

```python
plt.figure()
f, axarr = plt.subplots(1,5)
axarr[0].imshow(img_gray)
axarr[1].imshow(img_minmax)
axarr[2].imshow(img_luma)   
axarr[3].imshow(img_min)
axarr[4].imshow(img_max)

for i in range(5):
    axarr[i].set_yticklabels([])
    axarr[i].set_xticklabels([])

plt.show()
```

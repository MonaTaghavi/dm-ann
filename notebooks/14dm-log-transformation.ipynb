Random Skewed Distribution

```python
#from sklearn.datasets import make_regression
#generate a random distribution
X, y = make_regression(n_samples = 10000, noise = 100, random_state = 42
)
#convert the distribution to a skewed distribution
y = np.expm1((y+abs(y.min()))/200)

X, y
```

```python
fig = px.histogram(y)
fig.show()
```

Transform Skewed Data

```python
# np.log1p = log_n(x+1)
#log(0) # -inf

y_transformed = np.log1p(y)
```

```python
fig = px.histogram(y_transformed)
fig.show()
```

```python
y_back = np.expm1(y_transformed)
fig = px.histogram(y_back)
fig.show()
```
